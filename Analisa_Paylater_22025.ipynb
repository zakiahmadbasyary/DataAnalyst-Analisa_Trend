```python
# Membuat bar chart horizontal
fig = px.bar(
    df_fintech,
    x='Persen',
    y='Layanan',
    orientation='h',
    text='Persen',
    title='Penggunaan Layanan Fintech (%)',
    color='Persen',
     color_continuous_scale=[
        '#91C8E4',  # steelblue
        '#749BC2',  # dodgerblue
        '#3674B5'   # darkblue
    ]
)

# Atur tampilan teks
fig.update_traces(textposition='outside')

# Atur layout
fig.update_layout(
    title={
        'text': 'Penggunaan Layanan Fintech (%)',
        'y':0.95,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'
    },
    xaxis_title='Persentase',
    yaxis_title='Layanan',
    yaxis=dict(autorange="reversed"),
    height=400,
    width=1000,
)

fig.show()
```

## Pinjaman Konsumen Gopay

```python
df_gopay
```

```python
df_gopay.info()
```

```python
# Buat kolom label periode (misal: Q1 2024)
df_gopay['PeriodeLabel'] = df_gopay['Periode'] + ' ' + df_gopay['tahun'].astype(str)

# Gunakan PeriodeLabel sebagai sumbu X
fig = go.Figure()
fig.add_trace(go.Bar(
    x=df_gopay['PeriodeLabel'],
    y=df_gopay['Nilai (Triliun)'],
    marker_color='lightblue',
    text=df_gopay['Nilai (Triliun)'],
    textposition='outside',
    textfont=dict(color='black'),
    showlegend=False
))

# Line chart (tanpa label, tanpa legend)
fig.add_trace(go.Scatter(
    x=df_gopay['PeriodeLabel'],
    y=df_gopay['Nilai (Triliun)'],
    mode='lines+markers',
    line=dict(color='blue', width=3),
    showlegend=False
))

# Layout & Judul
fig.update_layout(
    title={
        'text': 'Perkembangan Pinjaman Konsumen Gopay Kuartalan (2024–2025)',
        'x': 0.5,
        'xanchor': 'center',
        'yanchor': 'top'
    },
    xaxis_title='Periode',
    yaxis_title='Nilai (Triliun)',
    plot_bgcolor='white',
    paper_bgcolor='white',
    height=500,
    width=800,
    showlegend=False
)

fig.show()
```
