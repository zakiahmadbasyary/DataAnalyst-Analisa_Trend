# Cleansing dan Preprocessing

```python
# menghapus kolom no dan mengubah tipe data kolom Hasil Survey ke numeric

def bersihkan_data(df):
    # 1. Hapus kolom 'No' (jika ada)
    if 'No' in df.columns:
        df = df.drop(columns=['No'])

    # 2. Bersihkan kolom 'Hasil Survei'
    if 'Hasil Survei' in df.columns:
        df['Hasil Survei'] = (
            df['Hasil Survei']
            .astype(str)                         # pastikan string
            .str.replace('%', '', regex=False)   # hilangkan %
            .str.replace(',', '.', regex=False)  # ganti koma → titik
            .apply(lambda x: re.sub(r'[^\d\.]', '', x))  # buang selain digit & titik
            .str.strip()
        )
        # 3. Ubah ke numerik
        df['Hasil Survei'] = pd.to_numeric(df['Hasil Survei'], errors='coerce')
    return df
```

```python
Data1=bersihkan_data(Data1)
```

```python
Data2=bersihkan_data(Data2)
```

# EDA

## Data 1

```python
# bar chart perbandingan  Cakupan DNKI
df_dnki = Data1[Data1['Metode'] == 'Cakupan DNKI']

#  Bersihkan kolom dan isi
df_dnki['Indeks'] = df_dnki['Indeks'].apply(lambda x: x.encode('ascii', 'ignore').decode('ascii').strip())
df_dnki = df_dnki.rename(columns={
    'Pekerjaan/ Kegiatan Sehari-hari': 'Pekerjaan',
    'Hasil Survei': 'Persentase'
})


# Buat clustered bar chart (Literasi vs Inklusi per Pekerjaan)
fig = px.bar(
    df_dnki,
    x='Pekerjaan',
    y='Persentase',
    color='Indeks',                     # Literasi & Inklusi
    category_orders={'Indeks': ['Literasi', 'Inklusi']},
    barmode='group',                   # side-by-side bar
    text_auto='.2f',
    labels={'Persentase': 'Persentase (%)'},
    title='Literasi vs Inklusi Keuangan per Kegiatan (Cakupan DNKI)'
)

fig.update_layout(
    xaxis_tickangle=-25,
    bargap=0.15,
    legend_title_text='Indeks',
    title_x=0.5
)
fig.update_traces(textfont_size=10)

fig.show()
```

```python
# Barchart Selisih
#Pivot: baris = Pekerjaan, kolom = Indeks -----------------------------
pivot = (
    df_dnki
    .pivot_table(index='Pekerjaan',
                 columns='Indeks',
                 values='Persentase')
    .reset_index()
)
pivot['Selisih'] = pivot['Inklusi'] - pivot['Literasi']
pivot = pivot.sort_values('Selisih', ascending=False)

fig = px.bar(
    pivot,
    x='Pekerjaan',
    y='Selisih',
    text='Selisih',
    labels={'Selisih': 'Selisih (Inklusi – Literasi) (%)'},
    title='Selisih Indeks Inklusi vs Literasi Keuangan per Kegiatan (Cakupan DNKI)',
    color='Selisih',            # otomatis gradasi; positif ≠ negatif jelas
    color_continuous_scale='Blues',  # merah‑biru (opsional)
)
fig.update_xaxes(categoryorder='array', categoryarray=pivot['Pekerjaan'])
fig.update_layout(
    xaxis_tickangle=-25,
    coloraxis_showscale=False,   # sembunyikan legenda warna
    bargap=0.15,
    title_x=0.5
)
fig.update_traces(texttemplate='%{text:.2f}', textfont_size=10)

fig.show()
```

```python
# Clean the 'Indeks' column to remove hidden characters
Data2['Indeks'] = Data2['Indeks'].apply(lambda x: x.encode('ascii', 'ignore').decode('ascii').strip())

# Display unique values in the 'Indeks' column
display(Data2['Indeks'].unique())
```

```python
# --- Pivot & selisih ---------------------------------------------------------
pivot3 = Data2.pivot_table(index='Lembaga',
                       columns='Indeks',
                       values='Persentase').reset_index()

pivot3['Selisih'] = pivot3['Literasi'] - pivot3['Inklusi']     # literasi – inklusi
pivot3['Tipe Gap'] = pivot3['Selisih'].apply(
    lambda x: 'Literasi>Mengungguli' if x > 0 else 'Inklusi>Mengungguli')

# Urutkan menurut |selisih| terbesar
pivot3 = pivot3.reindex(pivot3['Selisih'].abs().sort_values(ascending=False).index)


# --- Plot -------------------------------------------------------------------
import plotly.express as px
fig = px.bar(
    pivot3,
    x='Lembaga',
    y='Selisih',
    text='Selisih',
    color='Tipe Gap',
    color_discrete_map={
        'Literasi>Mengungguli':'#2ECC71',  # hijau
        'Inklusi>Mengungguli':'#E74C3C'    # merah
    },
    labels={'Selisih':'Literasi – Inklusi (%)'},
    title='Gap Literasi vs Inklusi per Sektor Jasa Keuangan'
)
fig.update_xaxes(categoryorder='array', categoryarray=pivot3['Lembaga'])
fig.update_layout(xaxis_tickangle=-25, bargap=0.2, title_x=0.5)
fig.update_traces(texttemplate='%{text:.2f}')

fig.show()
```
