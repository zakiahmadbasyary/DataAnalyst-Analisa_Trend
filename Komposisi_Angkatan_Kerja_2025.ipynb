# EDA

```python
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
import plotly.express as px

# 1️Buat list semua sektor
list_sektor = df['Lapangan Pekerjaan Utama'].unique()

# 2️Buat mapping warna
warna_sektor = {
    sektor: px.colors.qualitative.Plotly[i % len(px.colors.qualitative.Plotly)]
    for i, sektor in enumerate(list_sektor)
}

# 3️Loop dan tampilkan chart per tahun
tahun_list = df['Tahun'].unique()

for tahun in tahun_list:
    subset = df[df['Tahun'] == tahun]

    sektor_agg = subset.groupby('Lapangan Pekerjaan Utama')['Nilai'].sum().reset_index()

    fig = px.pie(sektor_agg,
                 names='Lapangan Pekerjaan Utama',
                 values='Nilai',
                 title=f'<b>Komposisi Angkatan Kerja per Sektor - Tahun {tahun}</b>',
                 hole=0.4,
                 color='Lapangan Pekerjaan Utama',
                 color_discrete_map=warna_sektor
                 )

    fig.update_layout(
        title_x=0.5,
        title_font=dict(size=20, family='Arial', color='black'),
        legend=dict(
            orientation="v",
            yanchor="middle",
            y=0.5,
            xanchor="left",
            x=1.05
        )
    )

    fig.show()
```

```python
# menampilkan presetase angkatan kerja per tahun dan dipisah bulan
# List tahun
tahun_list = df['Tahun'].unique()

# Loop per tahun
for tahun in tahun_list:
    # Filter data tahun tsb
    df_tahun = df[df['Tahun'] == tahun]

    # Data Februari
    df_feb = df_tahun[df_tahun['Bulan'] == 'Februari']
    sektor_feb = df_feb.groupby('Lapangan Pekerjaan Utama')['Nilai'].sum().reset_index()

    # Data Agustus
    df_ags = df_tahun[df_tahun['Bulan'] == 'Agustus']
    sektor_ags = df_ags.groupby('Lapangan Pekerjaan Utama')['Nilai'].sum().reset_index()

    # Buat subplot 1 row 2 kolom
    fig = make_subplots(rows=1, cols=2,
                        specs=[[{'type':'domain'}, {'type':'domain'}]],
                        subplot_titles=[f'Februari {tahun}', f'Agustus {tahun}'])

    # Pie chart Februari
    fig.add_trace(go.Pie(labels=sektor_feb['Lapangan Pekerjaan Utama'],
                         values=sektor_feb['Nilai'],
                         name='Februari'),
                  row=1, col=1)

    # Pie chart Agustus
    fig.add_trace(go.Pie(labels=sektor_ags['Lapangan Pekerjaan Utama'],
                         values=sektor_ags['Nilai'],
                         name='Agustus'),
                  row=1, col=2)

    # Layout
    fig.update_layout(
        title_text=f'<b>Komposisi Angkatan Kerja per Sektor - Tahun {tahun}</b>',
        title_x=0.5,
        title_font=dict(size=20, family='Arial', color='black'),
        legend=dict(
            orientation="v",
            yanchor="middle",
            y=0.5,
            xanchor="left",
            x=1.05
        ),
        height=600,
        width=2000
    )

    fig.show()
```

```python
# Line chart untuk per sektor dengan garis bulan dipisah
# Filter hanya Februari dan Agustus
df_filtered = df[df['Bulan'].isin(['Februari', 'Agustus'])]

# List sektor
sektor_list = df['Lapangan Pekerjaan Utama'].unique()

# Loop per sektor
for sektor in sektor_list:
    subset = df_filtered[df_filtered['Lapangan Pekerjaan Utama'] == sektor]

    fig = px.line(subset, x='Tahun', y='Nilai', color='Bulan',
                  title=f'Tren Nilai Februari & Agustus - {sektor}',
                  labels={'Tahun': 'Tahun', 'Nilai': 'Nilai', 'Bulan': 'Bulan'},
                  markers=True)

    fig.update_layout(
        title_x=0.5,
        title_font=dict(size=20, family='Arial', color='black'),
        xaxis_tickformat='d'  # format Tahun (angka)
    )

    fig.show()
```

```python
# line chart persektor dengan 2 bulan dalam 1 garis
# List sektor
sektor_list = df['Lapangan Pekerjaan Utama'].unique()

# Loop 1 sektor 1 grafik besar
for sektor in sektor_list:
    subset = df[df['Lapangan Pekerjaan Utama'] == sektor]

    fig = px.line(subset, x='Date', y='Nilai',
                  title=f'Tren Nilai - {sektor}',
                  labels={'Date': 'Tanggal', 'Nilai': 'Nilai'})

    # Format x-axis: Bulan Tahun
    fig.update_xaxes(
        tickformat="%b %Y",
        tickangle=0
    )
    fig.update_layout(
      title_x=0.5,
      title_font=dict(size=20, family='Arial', color='black')
    )

    fig.show()
```

```python
# Filter hanya Februari dan Agustus
df_filtered = df[df['Bulan'].isin(['Februari', 'Agustus'])]

# List sektor
sektor_list = df['Lapangan Pekerjaan Utama'].unique()

# Loop: 1 sektor → 1 barchart
for sektor in sektor_list:
    subset = df_filtered[df_filtered['Lapangan Pekerjaan Utama'] == sektor]

    fig = px.bar(subset,
                 x='Tahun',
                 y='Nilai',
                 color='Bulan',  # supaya ada 2 bar per tahun
                 barmode='group',  # pisah batang
                 text='Nilai',
                 title=f'<b>Perkembangan Angkatan Kerja (Februari & Agustus) - {sektor}</b>',
                 labels={'Tahun': 'Tahun', 'Nilai': 'Nilai', 'Bulan': 'Bulan'}
                )

    fig.update_layout(
        title_x=0.5,
        height=500,
        width=1000,
        title_font=dict(size=20, family='Arial', color='black'),
        xaxis_tickformat='d'
    )

    fig.show()
```

```python
# Filter data tahun 2024
df_2024 = df[df['Tahun'] == 2024]

# Agregasi total per sektor
sektor_agg = df_2024.groupby('Lapangan Pekerjaan Utama')['Nilai'].sum().reset_index().sort_values('Nilai', ascending=False)

# Barchart tanpa legend
fig = px.bar(sektor_agg,
             x='Lapangan Pekerjaan Utama',
             y='Nilai',
             color='Lapangan Pekerjaan Utama',
             text='Nilai',
             title='<b>Komposisi Angkatan Kerja per Sektor - Tahun 2024</b>',
             labels={'Lapangan Pekerjaan Utama': 'Sektor', 'Nilai': 'Nilai'}
            )

fig.update_layout(
    title_x=0.5,
    height=600,
    width=1000,
    title_font=dict(size=20, family='Arial', color='black'),
    xaxis_tickangle=-45,
    xaxis=dict(showticklabels=False),  # x label disembunyikan
    showlegend=False  # legend disembunyikan
)

fig.show()
```
