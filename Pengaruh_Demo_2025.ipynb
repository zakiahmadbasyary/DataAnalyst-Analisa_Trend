```python
df_IHSG.info()
```

```python
df_IHSG
```

```python
# Cari 3 nilai tertinggi
top3 = df_IHSG.nlargest(3, "Close")

# Plot line chart
fig = px.line(df_IHSG, x="Date", y="Close")

# Tambahin marker untuk 3 nilai tertinggi
fig.add_scatter(x=top3["Date"], y=top3["Close"],
                mode="markers+text",
                marker=dict(color="red", size=10),
                text=[f"{val:.2f}" for val in top3["Close"]],
                textposition="top right",
                name="3 Tertinggi")

fig.update_layout(
    title={
        'text': 'Pergerakan IHSG Agustus 2025',
        'y':0.95,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'
    },
    xaxis_title="Tanggal",
    yaxis_title="Harga Penutupan (Close)",
    hovermode="x unified"
)

fig.show()
```

```python
# --- Ambil 3 volume terbesar ---
top3_vol = df_IHSG.nlargest(3, "volume")

# Plot barchart Volume
fig = px.bar(df_IHSG, x="Date", y="volume",
             title="Volume Transaksi IHSG Agustus 2025",
             labels={"Date": "Tanggal", "volume": "Volume Transaksi"})

# Biar semua tanggal muncul di sumbu X
fig.update_xaxes(
    tickmode="array",
    tickvals=df_IHSG["Date"],   # semua tanggal dari DataFrame
    tickformat="%d"            # tampilkan hanya tanggal (01, 02, dst.)
)

# Tambahkan scatter untuk menandai 3 volume terbesar
fig.add_scatter(x=top3_vol["Date"], y=top3_vol["volume"],
                mode="markers+text",
                marker=dict(color="red", size=10),
                text=[f"{int(v):,}" for v in top3_vol["volume"]],
                textposition="top center",
                name="3 Volume Tertinggi")

fig.show()
```

```python
# Buat barchart dengan warna custom
fig = px.bar(
    df_BBCA,
    x="date",
    y="Nilai",
    color="Transaksi",
    barmode="group",
    text="Nilai",
    title="Perbandingan Transaksi Buy vs Sell BBCA",
    labels={"date": "Tanggal", "Nilai": "Nilai Transaksi"},
    color_discrete_map={"Buy": "blue", "Sell": "red"}  # mapping warna
)

# Format label text
fig.update_traces(texttemplate='%{text:,}', textposition='outside')

# Format x-axis
fig.update_xaxes(
    tickmode="array",
    tickvals=df_BBCA["date"].unique(),
    tickformat="%d-%m",
)

fig.show()
```
